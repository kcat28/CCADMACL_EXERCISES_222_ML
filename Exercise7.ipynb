```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x))
client_basket
```

```python
client_basket = client_basket.reset_index()
client_basket = client_basket.drop(columns=['Date', 'Member_number'])
client_basket = list(client_basket["itemDescription"].apply(lambda x:x.split(",")))
```

### 1. Apply one-hot encoding

```python
## put your answer here
one_hot_transformer = TransactionEncoder()
one_hot_data = one_hot_transformer.fit(client_basket).transform(client_basket)
one_hot_df = pd.DataFrame(one_hot_data, columns=one_hot_transformer.columns_)
one_hot_df
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
## put your answer here
basket_sets = apriori(one_hot_df, min_support=0.005, use_colnames=True)
basket_sets.sort_values("support", ascending=False)
```

3. Display the association rules using the lift metric

```python
## put your answer here
rules = association_rules(basket_sets, metric="lift", min_threshold=0.3)
rules.sort_values("lift", ascending=False)
#
```
